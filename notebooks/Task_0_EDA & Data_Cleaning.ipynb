```python
# how many missing values exists or better still what is the % of missing values in the dataset?
def percent_missing(df):
    # to calcualte total number of cells in dataframe
    totalCells=np.product(df.shape)
    # to count number of missing values percloumns
    missingCount=df.isnull().sum()
    # to calculate total number of missing values
    totalMissing=missingCount.sum()
    # to calucalte percentage of missing values 
    print("The Telecom dataset contains",round(((totalMissing/totalCells)*100),2),"%","missing values.")
percent_missing(df)
```

```python
# Geting which column has missing values 
df.isnull().sum()
```

```python
# Check the percentage of missing values in each column

percent_missing = df.isnull().sum() * 100 / len(df)
missing_value_df = pd.DataFrame({'percent_missing': percent_missing})
print(missing_value_df)
```

## Missing values Handling
##### Based on the above code some columns have missing values it need the way to handle it

```python
# Remove bad columns and duplicated Dur. (ms) column
clean_data.drop_columns(bad_columns)
clean_data.drop_columns('Dur. (ms).1')
```

```python
# Remove rows with smaller null values

clean_data.drop_rows(['Bearer Id', 'IMSI'])
```

```python
#Fill null categorical columns with Mode 

columns = ['Last Location Name', 'IMEI', 'MSISDN/Number', 'Handset Type', 'Handset Manufacturer']
clean_data.fill_categorical_columns(columns)
```

```python
null_info = DataFrameInfo(clean_data.df)
null_info.get_null_counts()
```

```python
# Fill null numerical columns with mean and median based on the column skewness value

column_list = pd.isnull(clean_data.df).columns[pd.isnull(clean_data.df).sum() > 0].tolist()
clean_data.fill_numerical_column(column_list)
```
